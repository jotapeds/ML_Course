```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib. pyplot as plt
import pandas as pd
import numpy as np
```

```python
from pyod.models.knn import KNN
```

```python
credit_base = pd.read_csv('credit_risk_dataset.csv')
credit_base.drop(credit_base.columns[[2,3,4,5,7,9,10,11]], axis = 1, inplace=True)
credit_base.dropna(inplace=True)
```

```python
credit_base.head(1)
```

```python
detector = KNN()
detector.fit(credit_base.iloc [:,1:4])
```

```python
predict = detector.labels_
predict
```

```python
np.unique(predict, return_counts = True)
```

```python
predict_score = detector.decision_scores_
predict_score
```

```python
#exporting data

outliers = []
for i in range(len(predict)):
    if predict[i] == 1:
        outliers.append(i)
```

```python
print(outliers)
```

```python
outliers_list = credit_base.iloc[outliers,:]
outliers_list
```
