```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
from sklearn.preprocessing import StandardScaler
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
x_random, y_random = make_blobs(n_samples= 200, centers =5, random_state = 2)
```

```python
x_random
```

```python
y_random
```

```python
grath = px.scatter(x = x_random[:,0], y = x_random[:,1])
grath.show()
```

```python
kmeans_blobs = KMeans(n_clusters=5)
kmeans_blobs.fit(x_random)
```

```python
labels = kmeans_blobs.predict(x_random)
labels
```

```python
center_clusters = kmeans_blobs.cluster_centers_
center_clusters
```

```python
grath_1 = px.scatter(x =x_random[:,0], y = x_random[:,1], color = labels)
grath_2 = px.scatter(x =center_clusters[:,0], y = center_clusters[:,1], size = [2,2,2,2,2])
grath_3 = go.Figure(data = grath_1.data + grath_2.data)
grath_3.show()
```
