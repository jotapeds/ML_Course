```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib. pyplot as plt
import pandas as pd
import numpy as np
```

```python
house_df = pd.read_csv('house_prices.csv')
house_df
```

```python
house_df.describe()
```

```python
x_houses = house_df.iloc[:,3:19].values
x_houses
```

```python
y_houses = house_df.iloc[:,2].values
y_houses
```

```python
from sklearn.model_selection import train_test_split
x_houses_test, x_houses_train, y_houses_test, y_houses_train = train_test_split(x_houses, y_houses, train_size=0.3, random_state=0)
x_houses_test.shape, x_houses_train.shape
```

```python
from sklearn.linear_model import ElasticNet
regr = ElasticNet(l1_ratio=0.001)
```

```python
houses_train = regr.fit(x_houses_train,y_houses_train)
```

```python
houses_train.score(x_houses_train,y_houses_train)
```

```python
predict = regr.predict(x_houses_test)
predict
```

```python
y_houses_test
```

```python
from sklearn.model_selection import cross_validate
```

```python
scores = cross_validate(regr, x_houses, y_houses, cv = 5, return_train_score=True,scoring= 'neg_mean_squared_error')
```

```python
print(sum(scores["train_score"]/5))
```
