```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib. pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
house_df = pd.read_csv('house_prices.csv')
house_df
```

```python
house_df.describe()
```

```python
house_df.isnull().sum()
```

```python

house_df.corr()
```

```python
figure = plt.figure(figsize=(20,20))
sns.heatmap(house_df.corr(), annot= True)
```

```python
#sqft_lot	
x_houses = house_df.iloc[:,5:6].values
x_houses
```

```python
#price
y_houses = house_df.iloc[:,2].values
y_houses
```

```python
from sklearn.model_selection import train_test_split
x_houses_test, x_houses_train, y_houses_test, y_houses_train = train_test_split(x_houses, y_houses, train_size=0.3, random_state=0)
```

```python
x_houses_train.shape, y_houses_train.shape
```

```python
x_houses_test.shape, y_houses_test.shape
```

```python
regressor_simple_houses = LinearRegression()
regressor_simple_houses.fit(x_houses_train, y_houses_train)
```

```python
regressor_simple_houses.intercept_
```

```python
regressor_simple_houses.coef_
```

```python
regressor_simple_houses.score(x_houses_train, y_houses_train)
```

```python
regressor_simple_houses.score(x_houses_test, y_houses_test)
```

```python
predict = regressor_simple_houses.predict(x_houses_train)
predict
```

```python
grath = px.scatter(x = x_houses_train.ravel(), y = predict)
grath.show()
```

```python
grath_1 = px.scatter(x = x_houses_train.ravel(), y = y_houses_train)
grath_2 = px.line(x = x_houses_train.ravel(), y = predict)
grath_2.data[0].line.color = 'red'

grath_3 =go.Figure(data=grath_1.data + grath_2.data )
grath_3
```

```python
predict_test = regressor_simple_houses.predict(x_houses_test)
predict_test
```

```python
y_houses_test
```

```python
abs(y_houses_test - predict_test).mean()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(y_houses_test,predict_test)
```

```python
mean_squared_error(y_houses_test, predict_test)
```

```python
np.sqrt(mean_squared_error(y_houses_test, predict_test))
```
