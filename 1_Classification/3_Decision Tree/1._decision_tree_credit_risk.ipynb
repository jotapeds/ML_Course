```python
# Decision Tree

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.tree import DecisionTreeClassifier
```

```python
import pickle
with open('credit_risk_base.pkl', 'rb') as f:
    x_credit_risk, y_credit_risk = pickle.load(f)
```

```python
x_credit_risk
```

```python
y_credit_risk
```

```python
credit_risk_tree = DecisionTreeClassifier(criterion='entropy') 
credit_risk_tree.fit(x_credit_risk, y_credit_risk)#forecast parameter, class
```

```python
from sklearn import tree
predict_names = ['historia', 'divida', 'garantia', 'renda'] # setting titles for the feature names
figure, axes = plt.subplots(nrows=1, ncols=1, figsize=(15,15)) # using matplotlib to make the figure more readable
tree.plot_tree(credit_risk_tree, feature_names=predict_names, class_names=credit_risk_tree.classes_, filled=True);# ploting tree, using feature names and classes names. ; used to hide text
```

```python
# associating
# historia boa(0), divida alta(0), garantia nenhuma(1), renda> 35(2)
#historia ruim(2), divida alta(0), garantia adequada(0), renda <15(0)

predict = credit_risk_tree.predict([[0,0,1,2],[2,0,0,0]])
predict
```
